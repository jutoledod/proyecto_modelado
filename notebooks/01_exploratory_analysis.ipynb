# Fase 1: Exploratory Data Analysis (EDA)

Este notebook realiza análisis exploratorio completo de los datos médicos.

## Objetivos:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import warnings

warnings.filterwarnings('ignore')

# Configuración de visualización
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.dpi'] = 100

# Seed para reproducibilidad
np.random.seed(42)
RANDOM_STATE = 42

# Añadir ruta src
sys.path.append('../src')

print("✅ Librerías importadas correctamente")
```

## 1. Carga de Datos

```python
# Cargar datos crudos
consultas = pd.read_csv('../data/consultas.csv')
examenes = pd.read_csv('../data/examenes.csv')
medicamentos = pd.read_csv('../data/medicamentos.csv')
pacientes = pd.read_csv('../data/pacientes.csv')

print("="*60)
print("DATOS CARGADOS - TAMAÑO ORIGINAL")
print("="*60)
print(f"Consultas:    {consultas.shape[0]} filas, {consultas.shape[1]} columnas")
print(f"Exámenes:     {examenes.shape[0]} filas, {examenes.shape[1]} columnas")
print(f"Medicamentos: {medicamentos.shape[0]} filas, {medicamentos.shape[1]} columnas")
print(f"Pacientes:    {pacientes.shape[0]} filas, {pacientes.shape[1]} columnas")
print(f"\nTotal de registros crudos: {consultas.shape[0] + examenes.shape[0] + medicamentos.shape[0] + pacientes.shape[0]}")
```

## 2. Exploración Inicial

### Información General

```python
# Información general - CONSULTAS
print("\n" + "="*60)
print("DATASET: CONSULTAS")
print("="*60)
print(consultas.info())
print("\nPrimeras 5 filas:")
print(consultas.head())
```

```python
# Información general - EXÁMENES
print("\n" + "="*60)
print("DATASET: EXÁMENES")
print("="*60)
print(examenes.info())
print("\nPrimeras 5 filas:")
print(examenes.head())
```

```python
# Información general - MEDICAMENTOS
print("\n" + "="*60)
print("DATASET: MEDICAMENTOS")
print("="*60)
print(medicamentos.info())
print("\nPrimeras 5 filas:")
print(medicamentos.head())
```

```python
# Información general - PACIENTES
print("\n" + "="*60)
print("DATASET: PACIENTES")
print("="*60)
print(pacientes.info())
print("\nPrimeras 5 filas:")
print(pacientes.head())
```

## 3. Estadísticas Descriptivas

```python
# Análisis de Valores Nulos (Por Dataset)
print("\n" + "="*60)
print("ANÁLISIS DE VALORES NULOS")
print("="*60)

datasets = {'Consultas': consultas, 'Exámenes': examenes, 
            'Medicamentos': medicamentos, 'Pacientes': pacientes}

for name, df in datasets.items():
    total_nulos = df.isnull().sum().sum()
    pct_nulos = (total_nulos / (df.shape[0] * df.shape[1])) * 100
    print(f"\n{name}: {total_nulos} valores nulos ({pct_nulos:.2f}%)")
    print(df.isnull().sum())
```

```python
# LIMPIEZA DE DATOS - Aplicando criterios de Eval 1
print("\n" + "="*60)
print("APLICANDO LIMPIEZA (CRITERIOS EVAL 1)")
print("="*60)

# 1. Eliminar filas con valores nulos (< 5% threshold)
print("\n1️⃣ Eliminando filas con valores nulos...")
consultas_clean = consultas.dropna()
examenes_clean = examenes.dropna()
medicamentos_clean = medicamentos.dropna()
pacientes_clean = pacientes.dropna()

print(f"Consultas:    {consultas.shape[0]} → {consultas_clean.shape[0]} filas ({(consultas_clean.shape[0]/consultas.shape[0]*100):.1f}%)")
print(f"Exámenes:     {examenes.shape[0]} → {examenes_clean.shape[0]} filas ({(examenes_clean.shape[0]/examenes.shape[0]*100):.1f}%)")
print(f"Medicamentos: {medicamentos.shape[0]} → {medicamentos_clean.shape[0]} filas ({(medicamentos_clean.shape[0]/medicamentos.shape[0]*100):.1f}%)")
print(f"Pacientes:    {pacientes.shape[0]} → {pacientes_clean.shape[0]} filas ({(pacientes_clean.shape[0]/pacientes.shape[0]*100):.1f}%)")

# 2. Eliminar duplicados
print("\n2️⃣ Eliminando duplicados...")
consultas_clean = consultas_clean.drop_duplicates()
examenes_clean = examenes_clean.drop_duplicates()
medicamentos_clean = medicamentos_clean.drop_duplicates()
pacientes_clean = pacientes_clean.drop_duplicates()

print(f"Consultas:    Duplicados removidos")
print(f"Exámenes:     Duplicados removidos")
print(f"Medicamentos: Duplicados removidos")
print(f"Pacientes:    Duplicados removidos")

# 3. Resetear índices
for df in [consultas_clean, examenes_clean, medicamentos_clean, pacientes_clean]:
    df.reset_index(drop=True, inplace=True)

print("\n✅ Limpieza completada")
print(f"\nTotal filas después de limpieza: {consultas_clean.shape[0] + examenes_clean.shape[0] + medicamentos_clean.shape[0] + pacientes_clean.shape[0]}")
```

```python
print("\n" + "="*60)
print("NORMALIZACIÓN DE CATEGORÍAS CATEGÓRICAS")
print("="*60)

# Normalizar especialidad (lowercase + strip)
if 'especialidad' in consultas_clean.columns:
    consultas_clean['especialidad'] = consultas_clean['especialidad'].str.lower().str.strip()
    print(f"✅ Especialidades normalizadas: {consultas_clean['especialidad'].nunique()} clases")

# Normalizar previsión (lowercase + strip)
if 'prevision' in pacientes_clean.columns:
    pacientes_clean['prevision'] = pacientes_clean['prevision'].str.lower().str.strip()
    print(f"✅ Previsiones normalizadas: {pacientes_clean['prevision'].nunique()} clases")
```

## 5. Decisión de Variable Objetivo (Target)

Para el modelado supervisado, necesitamos elegir qué predecir. Analizaremos opciones viables:

### Opción 1: Clasificación - ESPECIALIDAD MÉDICA
- **Target:** `especialidad` (de consultas)
- **Tipo:** Clasificación multiclase
- **Justificación:** La especialidad médica es un parámetro importante que puede predecirse según otras características

### Opción 2: Clasificación - PREVENCIÓN (Pacientes)
- **Target:** `prevision` (de pacientes)
- **Tipo:** Clasificación multiclase
- **Justificación:** La previsión de salud es relevante y clasificable

### Opción 3: Regresión - COSTO
- **Target:** `costo` (de consultas)
- **Tipo:** Regresión
- **Justificación:** Predecir costos es un problema de negocio común

Analizaremos la distribución de cada opción:

```python
# Análisis de opciones de TARGET

print("\n" + "="*60)
print("OPCIÓN 1: ESPECIALIDAD MÉDICA (Clasificación)")
print("="*60)
print("Distribución:")
print(consultas_clean['especialidad'].value_counts())
print(f"Clases únicas: {consultas_clean['especialidad'].nunique()}")

print("\n" + "="*60)
print("OPCIÓN 2: PREVENCIÓN (Clasificación)")
print("="*60)
print("Distribución:")
print(pacientes_clean['prevision'].value_counts())
print(f"Clases únicas: {pacientes_clean['prevision'].nunique()}")

print("\n" + "="*60)
print("OPCIÓN 3: TIPO DE MEDICAMENTO (Clasificación)")
print("="*60)
if 'tipo_medicamento' in medicamentos_clean.columns:
    print("Distribución:")
    print(medicamentos_clean['tipo_medicamento'].value_counts())
    print(f"Clases únicas: {medicamentos_clean['tipo_medicamento'].nunique()}")
else:
    print("Columna 'tipo_medicamento' no disponible en medicamentos_clean")
    print("Columnas disponibles:", medicamentos_clean.columns.tolist())
```
